<span style="font-size:30px; color:gray;"> Data balance 
<span style="font-size:20px; color:gray;">(Down sample)

```python

file_path = r"C:\Users\14217\Desktop\DM2583\project\data_set\Dataset_no1.csv"
df = pd.read_csv(file_path)

df_0 = df[df['Diabetes_012'] == 0.0]
df_2 = df[df['Diabetes_012'] == 2.0]

df_0_downsampled = df_0.sample(n=len(df_2), random_state=42)

df_balanced = pd.concat([df_0_downsampled, df_2]).sample(frac=1, random_state=42).reset_index(drop=True)

print("after balance categories distribute：")
print(df_balanced['Diabetes_012'].value_counts())

output_path = r"C:\Users\14217\Desktop\DM2583\project\data_set\Dataset_balanced.csv"
df_balanced.to_csv(output_path, index=False)

print(f"have generated balance dataset and save to: {output_path}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# ===================== Read Data =====================
file_raw = r"C:\Users\14217\Desktop\DM2583\project\data_set\Dataset_age_filtered.csv"
file_no1 = r"C:\Users\14217\Desktop\DM2583\project\data_set\Dataset_no1.csv"
file_bal = r"C:\Users\14217\Desktop\DM2583\project\data_set\Dataset_balanced.csv"

df_raw = pd.read_csv(file_raw)
df_no1 = pd.read_csv(file_no1)
df_bal = pd.read_csv(file_bal)

# ===================== Count Categories =====================
def get_counts(df):
    counts = df['Diabetes_012'].value_counts().sort_index()
    for cat in [0,1,2]:
        if cat not in counts:
            counts[cat] = 0
    return counts.sort_index()

counts_raw = get_counts(df_raw)
counts_no1 = get_counts(df_no1)
counts_bal = get_counts(df_bal)

# ===================== Combine for Plot =====================
counts_df = pd.DataFrame({
    'Original': counts_raw,
    'After removing 1': counts_no1,
    'After downsampling': counts_bal
})
counts_df = counts_df.T  # rows: stages, columns: categories

# ===================== Plot Stacked Bar Chart with Values =====================
categories = [0,1,2]
colors_map = {0:'blue', 1:'orange', 2:'red'}  # Blue:0, Orange:1, Red:2

plt.figure(figsize=(10,6))
bottom = np.zeros(len(counts_df))

for cat in categories:
    values = counts_df[cat]
    bars = plt.bar(counts_df.index, values, bottom=bottom, 
                   color=colors_map[cat], label=f'Class {cat}', alpha=0.8, edgecolor='black')
    
    # Add number labels on each segment
    for i, (b, val) in enumerate(zip(bottom, values)):
        if val > 0:  # only label if value > 0
            plt.text(i, b + val/2, str(val), ha='center', va='center', color='white', fontsize=10, fontweight='bold')
    
    bottom += values

plt.title("Diabetes_012 Category Distribution Across Stages", fontsize=14)
plt.xlabel("Data Stage")
plt.ylabel("Number of Samples")
plt.legend(title="Class", loc='upper right')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

<span style="font-size:30px; color:gray;"> data mapping
| feature      | mapping way           | mapping range | large value mean    |
| -------- | -------------- | ---- | -------- |
| MentHlth | Medical risk segment mapping       | 0–1  | ment healthy more seriously |
| PhysHlth | medical risk segment mapping       | 0–1  | physical healthy more seriously |
| BMI      | medical risk segment mapping       | 0–1  | weight healthy more seriously |
| GenHlth  |  1–5 → 0–1 | 0–1  | self healthy evaluate more bad  |



```python
import pandas as pd


file_path = r"C:\Users\14217\Desktop\DM2583\project\data_set\Dataset_reordered.csv"
df = pd.read_csv(file_path)


def health_risk_medical(x):
    """
    MentHlth / PhysHlth 风险映射，0-1
    """
    if x <= 5:
        return round(0 + (x / 5) * 0.2, 2)        # 0–0.2
    elif x <= 10:
        return round(0.2 + ((x - 5) / 5) * 0.2, 2)  # 0.2–0.4
    elif x <= 15:
        return round(0.4 + ((x - 10) / 5) * 0.2, 2) # 0.4–0.6
    elif x <= 20:
        return round(0.6 + ((x - 15) / 5) * 0.2, 2) # 0.6–0.8
    else:
        return round(0.8 + ((x - 20) / 10) * 0.2, 2) # 0.8–1.0

df['MentHlth'] = df['MentHlth'].apply(health_risk_medical)
df['PhysHlth'] = df['PhysHlth'].apply(health_risk_medical)


df['GenHlth'] = ((df['GenHlth'] - 1) / 4).round(1)


def map_bmi_medical(bmi):
    if bmi < 18.5:
        return round(0.2 + (bmi - 15) / (18.5 - 15) * 0.1, 2)  # 0.2-0.3
    elif bmi < 25:
        return round(0.0 + (bmi - 18.5) / (25 - 18.5) * 0.3, 2)  # 0.0-0.3
    elif bmi < 30:
        return round(0.3 + (bmi - 25) / (30 - 25) * 0.3, 2)  # 0.3-0.6
    elif bmi < 35:
        return round(0.6 + (bmi - 30) / (35 - 30) * 0.2, 2)  # 0.6-0.8
    else:
        return round(0.8 + min(0.2, (bmi - 35) / (45 - 35) * 0.2), 2)  # 0.8-1.0

df['BMI'] = df['BMI'].apply(map_bmi_medical)


print(df[['BMI', 'GenHlth', 'MentHlth', 'PhysHlth']].head())


output_path = r"C:\Users\14217\Desktop\DM2583\project\data_set\Dataset_mapped_medical.csv"
df.to_csv(output_path, index=False)
print(f"\nfinish medical risk segment mapping, and save to:{output_path}")
```
